```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("best_songs.csv",encoding='latin-1')
titles = df["title"]
artists = df["artist"]
year = df["year"]
```

```python
NB_TO_SHOW = 10

unq_artists,counts_artists = np.unique(artists.values,return_counts=True)
unq_and_counts = [(ar,c) for ar,c in zip(unq_artists,counts_artists)]
unq_and_counts_sorted = sorted(unq_and_counts,key=lambda x : x[1],reverse=True)
unq_artists_sorted = [tp[0] for tp in unq_and_counts_sorted][:NB_TO_SHOW]
unq_counts_sorted = [tp[1] for tp in unq_and_counts_sorted][:NB_TO_SHOW]

plt.figure(figsize=(15,10))
plt.bar(unq_artists_sorted,unq_counts_sorted)
plt.show()
```

```python
import re

words = []
for title in titles:
    pattern = re.compile(r'\([^)]*\)')

    title = re.sub(pattern, '', title)
    words.extend(title.split())

unq_words,unq_counts_words = np.unique(words,return_counts=True)
unq_and_counts = [(ar,c) for ar,c in zip(unq_words,unq_counts_words)]
unq_and_counts_sorted = sorted(unq_and_counts,key=lambda x : x[1],reverse=True)
unq_artists_sorted = [tp[0] for tp in unq_and_counts_sorted][:NB_TO_SHOW]
unq_counts_sorted = [tp[1] for tp in unq_and_counts_sorted][:NB_TO_SHOW]

plt.figure(figsize=(15,10))
plt.bar(unq_artists_sorted,unq_counts_sorted)
plt.show()
```
